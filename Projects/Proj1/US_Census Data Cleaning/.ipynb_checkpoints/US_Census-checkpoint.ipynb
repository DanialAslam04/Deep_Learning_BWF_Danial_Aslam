```python
import numpy as np
import pandas as pd
import glob as gb
import matplotlib.pyplot as plt
import seaborn as sns
```

## Reading CSV files

```python
csv_files = gb.glob( "*.csv")
csv_files
```

```python
dfs = []
for filename in csv_files:
    dff = pd.read_csv(filename)
    dfs.append(dff)

df = pd.concat(dfs,axis =0 ,ignore_index=True)
df
```

```python
df.shape
```

```python
df.columns
```

###### Dropping Unnamed

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.head()
```

#### Columns

```python
df.columns
```

#### Types

```python
df.dtypes
```

### Changing Types 

```python
per_col = ['Hispanic', 'White', 'Black', 'Native', 'Asian',
       'Pacific']
```

```python
for i in per_col:
    df[i] = df[i].str.rstrip('%').astype(float)
```

```python
df.head()
```

```python
df['Income'] = df['Income'].str.lstrip('$').astype(float)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df[['Males','Females']] = df['GenderPop'].str.split(pat='_', expand=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Males'].value_counts().sum()
```

```python
df['Males'] = df['Males'].astype(str).str.rstrip('M').astype(int)

df.head()
```

```python
df.dtypes
```

```python
df['Females'] = df['Females'].str.rstrip('F').replace('', np.nan).astype(float)
df.head()
```

### Missing values

```python
df['Females'].isnull().sum()
```

```python
df['Females'] = df['Females'].fillna(df['TotalPop']-df['Males'])
```

```python
df['Females'].isnull().sum()
```

### Duplicate values

```python
dup = df.duplicated().sum()
dup
```

```python
df = df.drop_duplicates()
```

```python
dup = df.duplicated().sum()
dup
```

```python
df.isnull().sum()
```

#### Missing values

```python
null = df[df['Pacific'].isnull()].index.tolist()
null
```

```python
df['Pacific'] = df['Pacific'].fillna(0)
df['Pacific'].isnull().sum()
```

#### Dropping unnecesary columns

```python
df = df.drop('GenderPop',axis =1)
```

```python
df.head()
```

### Scatter Plot

```python
sns.scatterplot(x=df['Females']/df['TotalPop'],y='Income',data= df)
```

### Histograms

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10, 10))

# Create a list of column names to plot
columns = per_col

# Use a for loop to plot histograms for each column of data
for i, ax in enumerate(axes.flat):
    if i < len(columns):
        sns.histplot(data=df, x=columns[i], ax=ax)
        ax.set_title(columns[i])

# Add vertical spacing between the subplots
plt.subplots_adjust(hspace=0.5)

# Add a main title to the entire figure
plt.suptitle('Histograms of Different Columns', fontsize=16)
```
