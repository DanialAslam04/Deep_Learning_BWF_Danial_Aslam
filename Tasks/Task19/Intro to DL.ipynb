### Deep Learning:
###### Deep Learning is a subfield of machine learning concerned with algorithms inspired by the structure and function of the brain called artificial neural networks. ANNs are composed of many interconnected layers of artificial neurons, to learn from large amounts of data. It involves building complex models that can automatically learn to recognize patterns and features in the data. 

### Neurons: 
###### Neurons are the basic computational units of an ANN. They receive inputs from other neurons or from the external environment, apply some computation to the inputs, and produce an output signal. The output signal may then be passed on to other neurons, or it may be the final output of the network.

### Neural Networks: 
###### A neural network is a series of algorithms that is used to recognize patterns in data. It is inspired by the structure and function of the human brain and consists of layers of interconnected nodes(neurons) that perform mathematical computations.

### Weights: 
###### Weights are the parameters of an ANN that are learned during training. Each neuron in the network has a set of weights associated with it, which determine how strongly it responds to the inputs it receives. During training, the weights are adjusted to minimize the error between the network's predictions and the true outputs.

### Layers: 
###### Layers are groups of neurons in an ANN that perform a specific type of computation. Input neurons receive data from the external environment, and output neurons produce the final output of the network. In between, there can be one or more layers of hidden neurons that perform intermediate computations. Each neuron in a layer receives inputs from the neurons in the previous layer, and its output is used as input for the neurons in the next layer.


### Backpropagation: 
###### Backpropagation is an algorithm used to train neural networks. It works by calculating the error between the predicted output of the network and the actual output and adjusting the weights of the connections between the nodes in the network to minimize this error.

### Activation functions: 
###### An activation function is a mathematical function that is applied to the output of a node in a neural network to introduce non-linearity into the network. This non-linearity allows the network to model more complex relationships in the data.

### Convolutional Neural Networks (CNNs): 
###### CNNs are a type of neural network that are designed for processing images and other spatial data. They use convolutional layers to extract features from the input data and pooling layers to reduce the size of the output.

### Dropout:
###### Dropout is a regularization technique used to prevent overfitting in neural networks. It works by randomly dropping out some of the nodes in the network during training, which forces the remaining nodes to learn more robust representations of the data.

### Loss Functions: 
###### A loss function is a mathematical function that measures the difference between the predicted output of a neural network and the actual output. It is used to guide the training of the network by providing feedback on how well it is performing.
